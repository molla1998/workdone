```python
data= pd.read_csv(r"C:\Users\TITAS\Downloads\P12-Churn-Modelling.csv")
```

```python
data.shape
```

```python
data_edited=data.drop(['RowNumber','CustomerId','Surname','Exited'],axis=1)
```

```python
data_edited.head(50)
```

```python
data_target=data['Exited']
```

```python
data_target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label1=LabelEncoder()
data_edited['Geography']=label1.fit_transform(data_edited['Geography'])
```

```python
label2=LabelEncoder()
data_edited['Gender']=label2.fit_transform(data_edited['Gender'])
```

```python
data_edited=pd.get_dummies(data_edited,drop_first=True,columns=['Geography'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(data_edited,data_target,test_size=0.1,random_state=0,stratify=data_target)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
y_train=y_train.to_numpy()
y_train
```

```python
x_train
```

```python
model=Sequential()
model.add(Dense(data_edited.shape[1],activation='relu',input_dim=data_edited.shape[1]))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
data_edited.shape[1]
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=10,epochs=10,verbose=1)
```

```python
y_pred=model.predict_classes(x_test)
```

```python
y_pred
```

```python
y_test=y_test.to_numpy()
```

```python
y_test
```

```python
model.evaluate(x_test,y_test)
```
